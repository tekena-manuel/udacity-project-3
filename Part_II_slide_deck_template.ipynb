## Distribution of User Type in Top 5 Stations

>The visualization below shows the top 5 busiest stations and it shows how much of its customers and subscribers make up the total trips. From the small subset chosen to perfomr this analysis it is clear that the subscribers make up the majority of trips taken in all the five stations



```python
#i will be making some categorical analysis: which users use the gobikes servies more
fig = plt.figure(figsize=(8,8))
plt.subplots_adjust(top=0.95)
fig.suptitle('RATIO OF USERS OF TOP 5 STATIONS' , fontsize=20)

sb.countplot(data = df, y = 'start_station_name', hue = 'user_type');
plt.xlabel('COUNT',fontsize=15)
plt.ylabel('STATION NAME',fontsize=15)
plt.legend(loc='center left', bbox_to_anchor=(1,0.5));
```

## Distribution of Gender in Top 5 Stations

> The visualization below shows the top 5 busiest stations and it reveals the gender ratio of its total trips. From the small subset of the dataframe chosen to perform this analysis it is clear that the males make up the majority of trips taken in all the five stations

```python
#i will be making some categorical analysis: which genders use the gobikes servies more
fig = plt.figure(figsize=(8,8))
plt.subplots_adjust(top=0.95)
fig.suptitle('Ratio of Gender usage' , fontsize=20)

sb.countplot(data = df, y = 'start_station_name', hue = 'member_gender')
plt.xlabel('COUNT',fontsize=15)
plt.ylabel('STATION NAME',fontsize=15)
plt.legend(loc='center left', bbox_to_anchor=(1,0.5));
```

## Distribution of Gender Types

> Now we will sort of combine visualization 2 and 3 to find out how much of the subcribers and customers are male and female. We come to realize that the mojority of both the customers and subcribers are male 

```python
#user type based on genders
sb.countplot(data=df, y='user_type', hue='member_gender')
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.title('Gender of User Types')
plt.xlabel('Count', fontsize = 15)
plt.ylabel('User Type', fontsize = 15);
```
